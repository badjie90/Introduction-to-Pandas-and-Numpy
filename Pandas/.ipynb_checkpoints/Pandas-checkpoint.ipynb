```python
car_sales_missing_data= pd.read_csv('car-sales-missing-data.csv')
car_sales_missing_data
```

in the above car-sales-missing-data.csv file, we have some
missing values represent by NaN. now we used the below 
code to replace the missing values with

```python
car_sales_missing_data['Odometer'].mean()
```

```python
# Now we will use the the mean value of the the odometer column to replace the missing 
# values with using a function called "fillna"
car_sales_missing_data['Odometer'].fillna(car_sales_missing_data['Odometer'].mean())
```

```python
car_sales_missing_data
```

If you can notice, our car-sales-missing-data.csv dataset sill have missing 
values which means, they are not replaced.
Now we are going to adjust our code be using "inplace=True"

```python
car_sales_missing_data['Odometer'].fillna(car_sales_missing_data['Odometer'].mean(), inplace=True)
```

```python
car_sales_missing_data
```
